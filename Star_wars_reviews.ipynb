```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head()
```

```python
star_wars.columns
```

## Converting column 2 & 3 into Binary Format

```python
yes_no = {"Yes": True, "No": False}

for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
    
```

```python
star_wars.columns[3:9]
```

```python
binary_mapping = {
    'Star Wars: Episode I  The Phantom Menace': True,
    np.nan: False,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(binary_mapping)
```

#### Renaming few columns where movie parts seen or not

```python
star_wars = star_wars.rename(columns={'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
    'Unnamed: 4': 'seen_2',
    'Unnamed: 5': 'seen_3',
    'Unnamed: 6': 'seen_4',
    'Unnamed: 7': 'seen_5',
    'Unnamed: 8': 'seen_6'})
star_wars.head(2)
```

# Rating (1to6) columns movie wise Data change to float and renaming

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.iloc[:,9:15].sample(2)
```

```python
star_wars = star_wars.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
                          'Unnamed: 10': 'ranking_2',
                          'Unnamed: 11': 'ranking_3',
                          'Unnamed: 12': 'ranking_4',
                          'Unnamed: 13': 'ranking_5',
                          'Unnamed: 14': 'ranking_6'
                         })
star_wars.iloc[:,9:15].head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
star_wars.iloc[:,9:15].mean().plot(kind='bar');
```

So far columns renamed and converted into binary format.  Rating 1 means most favourable and less favorable when rating number increases. Looks like part 5 is the most liked.

```python
star_wars.iloc[:,3:9].sum()
```

```python
star_wars.iloc[:,3:9].sum().plot.bar();
```

Viewer wise part 5 is top followed by 6 and 1 have most viewers in comparison to rest of them.

```python
male_viewer = star_wars[star_wars['Gender']== 'Male']
female_viewer = star_wars[star_wars['Gender']== 'Female']
```

```python
male_viewer.iloc[:,3:9].sum().plot.bar()
plt.title('Male Viewers')
plt.show()
female_viewer.iloc[:,3:9].sum().plot.bar()
plt.title('Female Viewers')
plt.show();
```

Overall male viewers are higher in comparison to female. however pattern of viewers movie wise is same among both the group such as part 5 is most viewed.

```python
male_viewer.iloc[:,9:15].mean().plot.bar()
plt.title('Male Rankings')
plt.show()
female_viewer.iloc[:,9:15].mean().plot.bar()
plt.title('Female Rankings')
plt.show();
```

Part 1 and 2 is more famous in woman in comparison to men while part 4 is less liked by woman in comparison to men. Ratings here are in reverse order as 1 mean better
